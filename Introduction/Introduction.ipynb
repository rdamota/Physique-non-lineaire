La dynamique de tous ces systèmes peuvent être modélisée par

$$\dot{\bf x} = f({\bf x}, t, \lambda)$$

où ${\bf x}$ est le <em>vecteur d'état</em> du système et $\lambda$ un paramètre de contrôle.

Si l'on a seulement une vue <em>stroboscopique</em> du système, sa dynamique peut alors être exprimée de la façon suivante

$${\bf x}^{(n+1)} = F({\bf x}^{(n)}, \lambda)$$

On a alors ce qu'on appelle un <em>itéré</em> ou une <em>map</em>.

<center><h2>La linéarité, kézako?</h2></center>

Considérons le système dynamique suivant

$$\dot{\bf x} = f(\bf x)$$

Sous quelles conditions ce système est-il <em>linéaire</em>?

Soit $u(t)$ et $v(t)$ deux solutions des équations. Le système est linéaire si:

- $w(t) = \alpha u(t)$ est aussi solution.
- $f(\alpha u + \beta v) = \alpha f(u) + \beta f(v)$
- La réponse est proportionelle à l'excitation.
- ...

Dans ce cas, notre système peut s'écrire sous la forme

$$\dot{\bf x} = {\bf Ax}$$

où ${\bf A}$ est un <em>opérateur linéaire</em> (en gros, une matrice).

<b>Exemple:</b> L'équation de l'oscillateur harmonique s'écrit

$$\ddot{x} = -\omega_0^2x$$

En posant $y = \dot{x}$, on obtient

$$ \frac{\mathrm{d}}{\mathrm{dt}}
\begin{bmatrix} 
    x \\
    y
\end{bmatrix} = \begin{bmatrix}
                    0 & 1 \\
                    -\omega_0^2 & 0
                \end{bmatrix} \begin{bmatrix}
                                x \\
                                y
                               \end{bmatrix}
$$

Un système linéaire obéit au <b>principe de supersition</b>, ce qui simplifie considérablement la résolution des problèmes.

<b>Exemple:</b> le champ magnétique généré par plusieurs conducteurs peut s’écrire comme étant la somme des champs générés par chacun des conducteurs pris isolément.

En apparence, la physique entière semble être gouvernée par des systèmes linéaires!

<h1><center><font color='red'>Attention!</font></center></h1>

Si les systèmes linéaires sont apparemment si fréquents, c’est uniquement parce qu’ils sont plus faciles à étudier! Beaucoup de systèmes d’apparence linéaire ne sont en réalité qu’une approximation d’un système non-linéaire plus complexe.

<h2><center>Le monde réel est non-linéaire!</center></h2>

Considérons le système dynamique suivant

$$\dot{\bf x} = f(\bf x)$$

Sous quelles conditions ce système <font color='red'>n</font>'est-il <font color='red'>pas</font> <em>linéaire</em>?

Soit $u(t)$ et $v(t)$ deux solutions des équations. Le système n'est pas linéaire si:

- $w(t) = \alpha u(t)$ n'est possible plus solution.
- $f(\alpha u + \beta v) \neq \alpha f(u) + \beta f(v)$
- La réponse n'est pas proportionelle à l'excitation.
- ...

<b>Exemple:</b> L'oscillateur de <em>Van der Pol</em>

$$\ddot{x} - \epsilon \omega_0 (1-x^2) \dot{x} + \omega_0^2 x = 0$$

permet de modéliser le potentiel d'action des neurones ou d'étudier l'intéraction des plaques au niveau d'une faille de subduction.

<b>Exemple:</b> L'équation du pendule oscillant est donnée par

$$\ddot{\theta} = -\omega_0^2 \sin (\theta)$$

Ce modèle est équivalent à l'oscillateur harmonique uniquement si $\vert \theta \vert \ll \pi/2$.

<b>Exemple:</b> L'émission de photons par un laser est donnée par

$$\dot{n} = gn(N_0-an) - kn$$

où $g$ est la gain du laser, $k$ décrit le taux de pertes et $N(t)=N_0-an$ est le nombre d'atomes excités.

Il est souvent illusoire de chercher à obtenir une solution analytique de ces équations, et une approche plus qualitative et <em>géométrique</em> du comportement global du système est préférable.

<center><h2>Illustration: le pendule oscillant</h2></center>

La dynamique d'un pendule oscillant amorti est donnée par

$$\ddot{\theta} = -2 \lambda \dot{\theta} - \omega_0^2 \sin (\theta)$$

En posant $\theta = u$ et $\dot{\theta} = v$, on a

$$
\begin{aligned}
& \dot{u} = v \\
& \dot{v} = -2 \lambda y - \omega_0^2 \sin(u)
\end{aligned}
$$

Traçons ce qu'on appelle <em>l'espace des phases</em> du système.

```octave
%--> Définition du maillage.
[u, v] = meshgrid(linspace(-2*pi, 2*pi, 25), linspace(-pi, pi, 25));

%--> Définition du système.
f = @(t, y) [y(2) ; -0*y(2)-sin(y(1))];

%--> Calcul de du/dt et dv/dt pour chacun des points du maillages.
udot = zeros(size(u)); vdot = zeros(size(v));
for i = 1:numel(u)
    yp = f(0, [u(i); v(i)]);
    udot(i) = yp(1);
    vdot(i) = yp(2);
end
```

```octave
quiver(u, v, udot, vdot, 'k'); xlabel('u'); ylabel('v'); axis equal tight;
```

Il existe certains points de cet espace des phases où le système semble ne rien faire. Ce sont des <em>points d'équilibre</em>. Ils sont donnés par

$$
\begin{aligned}
& \dot{u} = 0 \\
& \dot{v} = 0
\end{aligned}
$$

Dans notre cas, cela revient à

$$
\begin{aligned}
& v = 0
& \sin(u) = 0
\end{aligned}
$$

```octave
equilibres_u = [-2*pi, -pi, 0, pi, 2*pi]; equilibres_v = [0, 0, 0, 0, 0];
quiver(u, v, udot, vdot, 'k'); xlabel('u'); ylabel('v'); axis equal tight;
hold on;plot(equilibres_u, equilibres_v, 'r.', 'MarkerSize', 12);
```

```octave
%--> Trace le portrait de phase.
quiver(u, v, udot, vdot, 'k'); xlabel('u'); ylabel('v'); axis equal tight; hold on;plot(equilibres_u, equilibres_v, 'r.', 'MarkerSize', 12);
%--> Simule différentes trajectoires.
t = linspace(0, 50, 1000);
for v_init = [0, 0.5, 1, 1.5, 2, 2.5]
    y0 = [0; v_init];
    [ts, y] = ode45(f, t, y0); plot(y(:, 1), y(:, 2), 'LineWidth', 2.5);
end
```

L'espace des phases est structurés par ces points d'équilibres et par certaines trajectoirs qui semblent les relier.

Quid si on considère maintenant un amortissement non-nul?

```octave
%--> Définition du maillage.
[u, v] = meshgrid(linspace(-2*pi, 2*pi, 25), linspace(-pi, pi, 25));

%--> Définition du système.
f = @(t, y) [y(2) ; -0.1*y(2)-sin(y(1))];

%--> Calcul de du/dt et dv/dt pour chacun des points du maillages.
udot = zeros(size(u)); vdot = zeros(size(v));
for i = 1:numel(u)
    yp = f(0, [u(i); v(i)]);
    udot(i) = yp(1);
    vdot(i) = yp(2);
end
```

```octave
quiver(u, v, udot, vdot, 'k'); xlabel('u'); ylabel('v'); axis equal tight;
```

```octave
equilibres_u = [-2*pi, -pi, 0, pi, 2*pi]; equilibres_v = [0, 0, 0, 0, 0];
quiver(u, v, udot, vdot, 'k'); xlabel('u'); ylabel('v'); axis equal tight;
hold on;plot(equilibres_u, equilibres_v, 'r.', 'MarkerSize', 12);
```

```octave
%--> Trace le portrait de phase.
quiver(u, v, udot, vdot, 'k'); xlabel('u'); ylabel('v'); axis equal tight; hold on;plot(equilibres_u, equilibres_v, 'r.', 'MarkerSize', 12);
%--> Simule différentes trajectoires.
t = linspace(0, 50, 1000);
for v_init = [0, 0.5, 1, 1.5, 2, 2.5]
    y0 = [0; v_init];
    [ts, y] = ode45(f, t, y0); plot(y(:, 1), y(:, 2), 'LineWidth', 2.5);
end
```

<h2><center>Illustration: Le système de Lorenz</center></h2>

Le système de Lorenz est donné par

$$
\begin{aligned}
    & \dot{x} = \sigma (y-x) \\
    & \dot{y} = x(\rho -z) - y \\
    & \dot{z} = xy - \beta z.
\end{aligned}
$$

C'est un modèle simplifié de la convection atmosphérique dévelopé en 1963.

```octave
%--> Définition des paramètres du système.
sigma = 10; rho = 99.96 ; beta = 8/3;

%--> Définition du système dynamique.
f = @(t, y) [sigma*(y(2)-y(1)) ; y(1)*(rho - y(3))-y(2) ; y(1)*y(2) - beta*y(3)];

%--> Points d'équilibre du système.
eq_x = [0, sqrt(beta*(rho-1)), -sqrt(beta*(rho-1))];
eq_y = [0, sqrt(beta*(rho-1)), -sqrt(beta*(rho-1))];
eq_z = [0, rho-1, rho-1];

%--> Définition de la condition initiale.
y0_a = [0.0001, 0.0001, 0.0001];
y0_b = [0.0001, -0.0001, 0.0001];

%--> Simulation du système.
t = linspace(0, 100, 20000);
[ts, ya] = ode45(f, t, y0_a);
[ts, yb] = ode45(f, t, y0_b);
```

```octave
%--> Trace l'évolution temporelle des différentes composantes du système.
h1 = subplot(3, 1, 1); plot(ts, ya(:, 1), ts, yb(:, 1)); ylabel('x'); h1_pos = get(h1, 'Position'); set(gca, 'XTickLabel', "");
h2 = subplot(3, 1, 2); plot(ts, ya(:, 2), ts, yb(:, 2)); ylabel('y'); h2_pos = get(h2, 'Position'); set(h2, 'Position',[h1_pos(1) h1_pos(2)-1.2*h1_pos(4) h2_pos(3:end)]); set(gca, 'XTickLabel', "");
h3 = subplot(3, 1, 3); plot(ts, ya(:, 3), ts, yb(:, 3)); ylabel('z'); xlabel('t'); h3_pos = get(h3, 'Position'); set(h3, 'Position',[h2_pos(1) h2_pos(2)-.8*h1_pos(4) h3_pos(3:end)]); 
```

```octave
%--> Trajectoire dans l'espace des phases.
plot3(ya(:, 1), ya(:, 2), ya(:, 3)); hold on; plot3(yb(:, 1), yb(:, 2), yb(:, 3), '--');
plot3(eq_x, eq_y, eq_z, 'r.', 'MarkerSize', 12); xlabel('x'); ylabel('y'), zlabel('z');

%--> Change the view.
view(45, 30)
```

Malgré sa simplicité apparente, le système de Lorenz possède une dynamique extrêmement riche!

<center><h2>Conclusion</h2></center>

Pour comprendre la dynamique globale d'un système, il suffit donc de connaître

- les <em>attracteurs</em> du système et leur stabilité,
- comment ces attracteurs sont liés entre eux dans l'espace des phases,
- comment ces attracteurs et leur stabilité varient si on change le paramètre de contrôle $\lambda$.

Dans la suite de ce cours, nous verrons:

- comment calculer les points d'équilibre du système?
- comment étudier leur stabilité?
- comment étudier les bifurcations successives d'un système?
- comment caractériser le chaos déterministe?
