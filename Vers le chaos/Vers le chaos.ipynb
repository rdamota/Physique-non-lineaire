<center><h1>Vers le chaos</h1></center>

<p><center>Physique non-linéaire</center></p>
<p><center>Jean-Christophe Loiseau, MdC, Arts et Métiers ParisTech</center></p>

Soit un système dynamique dépendant d'un paramètre de contrôle $r$

$$\dot{\bf x} = f({\bf x}, r)$$

où $f$ est une fonction non-linéaire. Nous avons vu précédemment que la dynamique du système peut changer radicalement lorsque $r$ varie.

Considérons le système dit de <em>Rössler</em>

$$
\begin{align}
& \dot{x} = -y-z \\
& \dot{y} = x + ay \\
& \dot{z} = b + z(x-c)
\end{align}
$$

Décrit par Otto Rössler en 1976 dans un but purement théorique, ce système s'est avéré utiles dans la modélisation de l'équilibre dans des réactions chimiques.

```python
rossler = lambda x, t, a, b, c: np.array([-x[1]-x[2] , x[0] + a*x[1], b + x[2]*(x[0]-c)])
a = 0.2 ; b = 0.2 ; c = 4.5 ; t = np.linspace(0, 500, 15000) ; y0 = np.array([0.1, 0.1, 0.1])
y = odeint(rossler, y0, t, args=(a, b, c))
```

```python
time_series(t, y)
```

```python
phase_space(y)
```

En fonction de la valeur de $c$, on observe plusieurs régime:

- $c=4 \rightarrow$ cycle de période 1.
- $c=6 \rightarrow$ cycle de période 2.
- $c=8.5 \rightarrow$ cycle de période 4.
- $c=9 \rightarrow$ dynamique chaotique.

Intéressons nous plus en détails au cas où le système est chaotique et essayons de définir ce que l'on entend par là.

```python
c = 15 ; init_0 = np.array([0.1, 0.1, 0.1]) ; init_1 = np.array([0.11, 0.1, 0.1])
y0 = odeint(rossler, init_0, t, args=(a, b, c)) ; y1 = odeint(rossler, init_1, t, args=(a, b, c))
```

```python
compare_time_series(t, y0, y1)
```

```python
compare_time_series_bis(t, y0, y1)
```

Résumons:

- En augmentant $c$, le système subit différentes bifurcations successives.
- La dynamique du système finie par sembler désordonnée, imprévisible.
- À ce stade, deux conditions initiales extrêmement proches divergent exponentiellement.

<center><h2>Disgression sur la notion de hasard et de déterminisme</center></h2>

Les <em>systèmes chaotiques</em> s'oppossent aux <em>systèmes intégrables</em> de la mécanique classique (entre autres), qui furent longtemps les symboles d'une régularité toute puissante en physique théorique.

<center><h3>Le déterminisme Laplacien</center></h3>

*Nous devons [...] envisager l'état présent de l'univers comme l'effet de son état antérieur, et comme la cause de celui qui va suivre. Une intelligence qui pour un instant donné connaîtrait toutes les forces dont la nature est animée et la situation respective des êtres qui la composent [...] embrasserait dans la même formule les mouvements des plus grands corps de l'univers et ceux du plus léger atome : rien ne serait incertain pour elle, et l'avenir comme le passé serait présent à ses yeux.*

</br>

*Essai philosophique sur les probabilités*, Pierre-Simon de Laplace, 1814

Ce texte a trouvé un écho quelques années plus tard avec le théorème de Cauchy-Lipschitz.

<center><h3>Poincaré et l'imprédictibilité</h3></center>

*Une cause très petite, qui nous échappe, détermine un effet considérable que nous ne pouvons pas ne pas voir, et alors nous disons que cet effet est dû au hasard. Si nous connaissions exactement les lois de la nature et la situation de l'univers à l'instant initial, nous pourrions prédire exactement la situation de ce même univers à un instant ultérieur. Mais, lors même que les lois naturelles n'auraient plus de secret pour nous, nous ne pourrions connaître la situation qu'approximativement. Si cela nous permet de prévoir la situation ultérieure avec la même approximation, c'est tout ce qu'il nous faut, nous disons que le phénomène a été prévu, qu'il est régi par des lois ; mais il n'en est pas toujours ainsi, il peut arriver que de petites différences dans les conditions initiales en engendrent de très grandes dans les phénomènes finaux ; une petite erreur sur les premières produirait une erreur énorme sur les derniers. La prédiction devient impossible et nous avons le phénomène fortuit.*

</br>

*Calcul des Probabilités*, Henri Poincaré, 1912

Ce que décrit H. Poincaré est ce qu'on appelle aujourd'hui la *sensibilité aux conditions aux initiales*: pour un système chaotique, une très petite erreur sur la connaissance de l'état initial va se trouver (presque toujours) rapidement amplifiée.

<center><h2>Différentes routes vers le chaos</h2></center>

Il existe trois grandes routes par lesquelles un système peut transitioner vers le chaos:

- la cascade sous-harmonique
- le scénario de Ruelle-Takens
- les intermittences de Pomeau-Manneville.

<center><h3>Feigenbaum: Cascade de doublements de période</h3></center>

Considérons la suite logistique

$$
x_{n+1} = r x_n (1-x_n)
$$

où $n = 0, 1, \cdots$ dénote le temps discret, $x$ est l'unique degré de liberté du système et $0 \le r \le 4$ est le paramètre de contrôle.

```python
_, x = suite_logistique(3.6, niter=250)
plot_suite(x)
```

```python
r = np.linspace(0, 4, 2000)
bifurcation_diagram(r)
```

Après plusieurs doublement de périodes, le système présente une dynamique complètement chaotique.

Notez malgré tout qu'il existe certaines plages de $r$ pour lequelle le système retrouve un comportement périodique.

C'est ce scénario de transition vers le chaos qui a lieu pour le système de Rössler.

```python
a = 0.2 ; b = a ; y0 = np.array([0.1, 0.1, 0.1]) ; c = [2.8, 3.5, 4, 4.2] ; t = np.linspace(0, 1500, 15000)
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(6, 6))
for ii, ax in enumerate(axes.ravel()):
    y = odeint(rossler, y0, t, args=(a, b, c[ii])) ; ax.plot(y[-1000:, 0], y[-1000:, 1], lw=.5) ;ax.set_title('$c = %.1f$' %c[ii]) ; ax.locator_params(axis='both', nbins=5)
```

<center><h3>Ruelle-Takens: Quasi-périodicité</h3></center>

Scénario proposé en 1975 par David Ruelle et Floris Takens, il va à l'encontre de la théorie de Lev Landau acceptée jusqu'à présent.

<center><h3>Pomeau-Manneville : l'intermittence</h3></center>

L'intermittence à la Pomeau-Manneville est l'alternance irrégulière de phases apparement pérodiques et de dynamiques chaotiques : un système en apparence quasi-périodique peut présenter des *bursts* irréguliers. Au moins trois types différents d'intermittence ont été décrit:

- type I (saddle-node bifurcation)
- type II (subcritical Hopf bifurcation)
- type III (inverse period-doubling bifurcation)

<center><h3>Intermittence de type I</h3></center>

```python
lorenz = lambda x, t, sigma, r, b : np.array([sigma*(x[1]-x[0]) , -x[0]*x[2] + r*x[0] - x[1] , x[0]*x[1]-b*x[2]])
t = np.linspace(0, 75, 20000) ; y0 = np.array([50., 50.1, 50.1])
y = odeint(lorenz, y0, t, args=(10, 166.5, 8/3.))[1000:] ; t=t[1000:]
time_series(t, y)
```

```python
phase_space(y)
```
