<center><h2>Bifurcations pour des systèmes du premier ordre</h2></center>

</br>

<center><img src='Figures/diagramme_bifurcation.jpg' width=500px></center>

Commençons par considérer un système du premier ordre (un seul degré de liberté) donné par

$$\dot{x} = f(x, r)$$

où $r$ est notre paramètre de contrôle.

De tels systèmes ont une dynamique relativement simple entièrement caractérisée par les points fixes du dit-système.

Ces points fixes peuvent néanmoins changer qualitativement et quantitativement si les paramètres de contrôle varient.

- Les variations qualitatives de la dynamique sont appelées <em>bifurcations</em>.
- Les valeurs des paramètres de contrôle pour lesquelles ont lieu ces bifurcations sont appelés <em>points de bifurcation</em>.

Écrivons le dévelopement limité de $f$ autour du point fixe $x^*$

$$f(x) \simeq \alpha + \beta x + \gamma x^2 + \delta x^3$$

En fonction des valeurs de $\alpha$, $\beta$, $\gamma$ et $\delta$, différents comportements sont possibles.

<center><h3>La bifurcation noeud-col </h3></center>

Considérons tout d'abord le système donné par

$$\dot{x} = \mu - x^2$$

et traçons son portrait de phase pour différents $\mu$.

```python
f = lambda x, mu: mu - x**2
x = linspace(-2, 2)
phase_line_plot(x, f)
```

En fonction de la valeur de $\mu$, plusieurs comportements sont possibles:
    
- Pour $\mu < 0$, le système ne possède aucun point fixe.
- Pour $\mu > 0$, le système possède deux points fixes, $x^*=\pm \sqrt{\mu}$, dont l'un est stable et l'autre instable.

À $\mu = 0$, on observe une transition entre l'existence d'aucun point fixe à celle de deux points fixes, dont un est stable et l'autre instable. C'est ce qu'on appelle la bifurcation <em>noeud-col</em>.

```python
saddle_node_bifurcation()
```

<center><h3>La bifurcation transcritique</h3></center>

Considérons maintenant le système donné par

$$\dot{x} = \mu x - x^2$$

et traçons son portrait de phase.

```python
f = lambda x, mu: 3*mu*x - x**2
x = linspace(-2, 2)
phase_line_plot(x, f)
```

Les points fixes du système sont

$$x^*_1 = 0 \text{ et } x^*_2 = \mu$$

On peut montrer que

- pour $\mu < 0$, $x^*_1$ est stable tandis que $x^*_2$ est instable.
- pour $\mu > 0$, $x^*_1$ est instable tandis que $x^*_2$ est stable.

À  $\mu = 0$, les deux points fixes échangent leur stabilité. C'est ce qu'on appelle la bifurcation transcritique.

```python
transcritical_bifurcation()
```

<center><h3>La bifurcation fourche</h3></center>

Considérons le système donné par

$$\dot{x} = \mu x - x^3$$

et traçons son portrait de phase.

```python
f = lambda x, mu: mu*x + x**3
x = linspace(-2, 2)
phase_line_plot(x, f)
```

En fonction de la valeur de $\mu$, plusieurs comportements sont possibles (pour un terme cubique négatif):
    
- Pour $\mu < 0$, le système ne possède qu'un unique point fixe stable.
- Pour $\mu > 0$, le système possède maintenant trois points fixes, dont un instable et deux stables.

En fonction de la valeur de $\mu$, plusieurs comportements sont possibles (pour un terme cubique positif):
    
- Pour $\mu > 0$, le système ne possède qu'un unique point fixe instable.
- Pour $\mu < 0$, le système possède maintenant trois points fixes, dont deux instables et un stable.

```python
pitchfork_bifurcation()
```

Considérons maintenant le modèle suivant

$$\dot{x} = \mu x + x^3 - 0.25 x^5$$

Il s'agit du même modèle que la fourche sous-critique mais le dévelopement limité de $f(x)$ a été poussé à un ordre plus élevé.

```python
pitchfork_bifurcation_bis()
```

Le système présente maintenant trois bifurcations distinctes:
- une bifurcation fourche sous-critique à $\mu = 0$,
- deux bifurcations noeud-col à $\mu = 1$.

Le système présente également la possibilité d'hystérésis.

<center><h3>Les bifurcations fourche impropres</h3></center>

$$\dot{x} = \alpha + \mu x - x^3$$

Le terme constant modélise une imperfection qui fait préferer au système une des deux branches.

```python
pitchfork_bifurcation_impropre()
```

Le terme constant sert à "briser" la bifurcation de fourche : une des deux nouvelles branches devient une continuation de la branche de départ et l’autre branche se sépare. La bifurcation de fourche est transformée en bifurcation noeud-col.

<center><h2>Bifurcations pour des sytèmes du deuxième ordre</h2></center>

Nous avons vu lors du dernier cours que les systèmes du deuxième ordre peuvent présenter une dynamique oscillatoire correspondant à ce qu'on appelle un <em>cycle limite</em>.

Considérons le système suivant

$$
\begin{align}
& \dot{x} = \mu x - \omega y - (x^2+y^2)x \\
& \dot{y} = \omega x + \mu y -(x^2 + y^2)y
\end{align}
$$

qui possède un unique point fixe donné par $(x^*, y^*) = (0, 0)$.


```python
f = lambda x, y, mu : np.array([mu*x - y - (x**2+y**2)*x , x + mu*y - (x**2+y**2)*y])
x = linspace(-1, 1); x, y = meshgrid(x, x)
phase_plot_2d(x, y, f)
```

Tout comme précédemment, le système subit une bifurcation à $\mu = 0$:
- pour $\mu < 0$, la spirale est attractive.
- pour $\mu > 0$, la spirale est répulsive et on assite à la création d'un cycle limite.

Une telle bifurcation est appelée bifurcation de <em>Andronov-Poincaré-Hopf</em>.

En introduisant la variable complexe $z = x + iy$, le système précédent peut s'écrire

$$\dot{z} = (\mu + i\omega)z - \vert z \vert^2 z$$

Écrivons maintenant $z = r \mathrm{e}^{i \theta}$. On peut alors mettre notre système initial sous forme polaire

$$
\begin{align}
& \dot{r} = \mu r - r^3 \\
& \dot{\theta} = \omega
\end{align}
$$

où $r$ est l'amplitude de notre oscillation. Pour le système considéré, on voit que cette amplitude subit une bifurcation fourche supercritique à $\mu = 0$.

Il existe évidemment aussi des bifurcations de Hopf sous-critique dont une des formes normales est donnée par


$$
\begin{align}
& \dot{r} = \mu r + r^3 \\
& \dot{\theta} = \omega
\end{align}
$$

Notez par ailleurs que la bifurcation de Hopf n'est pas l'unique façon de créer des cycles limites et que les cycles limites eux-même peuvent subir des bifurcations.

<center><h2>Bifurcations globales</h2></center>

Les bifurcations vues jusqu'à maintenant correspondent à une modification <em>locale</em> de l'espace des phases. Elles peuvent par ailleurs être détectées grâce à une analyse de stabilité linéaire du système linéarisé autour du point fixe.

<center><h3>Bifurcation noeud-col de cycles limites</h3></center>

Considérons le système suivant (en coordonées polaires)

$$
\begin{align}
& \dot{r} = \mu r + r^3 - r^5 \\
& \dot{\theta} = \omega + br^2
\end{align}
$$

Il s'agit de la forme normale d'une bifurcation de Hopf sous-critique (pour $\mu = 0$). Intéressons nous plus particulièrement à ce qu'il se passe pour $\mu < 0$.

```python
saddle_node_of_periodic_orbits(mu=-0.15)
```

<center><h3>Bifurcation homocline</h3></center>

Considérons maintenant le système donné par

$$
\begin{align}
& \dot{x} = y \\
& \dot{y} = \mu y + x - x^2 + xy
\end{align}
$$

Il possède deux points fixes: $(x_1^*, y_1^*) = (0, 0)$ et $(x_2^*, y_2^*) = (1, 0)$.

Les matrices Jacobiennes associées à chacun de ces points fixes sont

$${\bf J}_1 = \begin{bmatrix}
                0 & 1 \\
                1 & \mu
              \end{bmatrix} \text{ et } {\bf J}_2 = \begin{bmatrix}
                                                      0 & 1 \\
                                                      -1 & \mu+1
                                                    \end{bmatrix}
$$

Pour $-1 <\mu < 0$, $(0, 0)$ correspond alors à un col, tandis que $(1, 0)$ est une spirale instable.

```python
mu = -0.92
f = lambda y, t: np.array([ y[1] , mu*y[1] + y[0] - y[0]**2 + y[0]*y[1] ])
phase_plot_homoclinic(f)
```

Ceci est une bifurcation <em>homocline</em>. Elle cause un changement topologique des trajectoires dans l'espace de phase alors que la nature des points fixes n'a pas changée.
